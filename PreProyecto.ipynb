# Identificar el objetivo target (Y)

Y : son los eventos de bancarrota 1:si y 0:para no

```python
Y = df['event']
conteo_clases = Y.value_counts()
print(conteo_clases)
```

Se observa que en el df tenemos 13850 empresas que no han caído en bancarrota y 247 que si cayeron en bancarrota

```python
plt.figure(figsize=(4,3))
conteo_clases.plot(kind='bar')
plt.title('Conteo de Clases')
plt.xlabel('Clase: Bancarrota')
plt.ylabel('Frecuencia')
plt.show()
```
